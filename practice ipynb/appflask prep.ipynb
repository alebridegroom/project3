```python
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
from sqlalchemy import Column, Integer, String, Float

from sqlalchemy.ext.declarative import declarative_base
# Create a base class for declarating class definitions to produce Table objects
Base = declarative_base()
```

```python
Base = declarative_base()
```

```python
class census(Base):
    __tablename__ = "census"
    State = Column(String)
    County = Column(String, primary_key=True)
    Women = Column(Float)
    White= Column(Float)
    Citizen = Column(Float)
    IncomePerCap = Column(Float)
    Poverty = Column(Float)
    ChildPoverty = Column(Float)
    Professional = Column(Float)
    Service = Column(Float)
    Office = Column(Float)
    Construction = Column(Float)
    Production = Column(Float)
    Drive = Column(Float)
    Carpool = Column(Float)
    Transit = Column(Float)
    Walk = Column(Float)
    OtherTransp = Column(Float)
    WorkAtHome = Column(Float)
    MeanCommute = Column(Float)
    Employed = Column(Float)
    PrivateWork = Column(Float)
    PublicWork = Column(Float)
    SelfEmployed= Column(Float)
    FamilyWork = Column(Float)
    Unemployment = Column(Float)
    Minority = Column(Float)
```

```python
class election(Base):
    __tablename__ = "election"
    ID = Column(Integer, primary_key=True)
    county = Column(String)
    fips = Column(Integer)
    candidate = Column(String)
    state = Column(String)
    votes = Column(Integer)
```

```python
class federal(Base):
    __tablename__ = "federal"
    ID = Column(Integer, primary_key=True)
    candidate = Column(String)
    votes = Column(Integer)
```

```python
class state(Base):
    __tablename__ = "state"
    ID = Column(Integer, primary_key=True)
    candidate = Column(String)
    state = Column(String)
    votes = Column(Integer)
```

```python
class winners(Base):
    __tablename__ = "winners"
    ID = Column(Integer, primary_key=True)
    county = Column(String)
    fips = Column(Integer)
    candidate = Column(String)
    state = Column(String)
    votes = Column(Integer)
    total = Column(Integer)
    pct = Column(Float)
    Women = Column(Float)
    White= Column(Float)
    Citizen = Column(Float)
    IncomePerCap = Column(Float)
    Poverty = Column(Float)
    ChildPoverty = Column(Float)
    Professional = Column(Float)
    Service = Column(Float)
    Office = Column(Float)
    Construction = Column(Float)
    Production = Column(Float)
    Drive = Column(Float)
    Carpool = Column(Float)
    Transit = Column(Float)
    Walk = Column(Float)
    OtherTransp = Column(Float)
    WorkAtHome = Column(Float)
    MeanCommute = Column(Float)
    Employed = Column(Float)
    PrivateWork = Column(Float)
    PublicWork = Column(Float)
    SelfEmployed= Column(Float)
    FamilyWork = Column(Float)
    Unemployment = Column(Float)
    Minority = Column(Float)
    mean_income_above_24K = Column(String)
```

```python
engine = create_engine("sqlite:///election.sqlite")
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(bind=engine)
```

```python
election_2016 = session.query(winners)
for player in election_2016:
    print(player.Unemployment)
```
