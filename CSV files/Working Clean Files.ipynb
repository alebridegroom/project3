```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Name of the CSV file
presidentcsv = Path('uncleaned/1976-2020-president.csv')
```

```python
# The correct encoding must be used to read the CSV in pandas
presidentdf = pd.read_csv(presidentcsv, encoding="UTF-8")
```

```python
presidentdf.head()
```

```python
cleanpresidentdf = presidentdf[['year', 'state', 'state_po', 'state_fips', 'state_cen', 'state_ic', 'office', 'candidate', 'party_detailed', 'writein', 'candidatevotes', 'totalvotes']]
cleanpresidentdf.head()
```

```python
yearfilter2016df = cleanpresidentdf[cleanpresidentdf['year'] == 2016]
```

```python
yearfilter2016df.head()
```

```python
yearfilter2016df.to_csv("cleaned/2016clean.csv", index=False, encoding="UTF-8")
```

```python
yearfilter2020df = cleanpresidentdf[cleanpresidentdf['year'] == 2020]
```

```python
yearfilter2020df.head()
```

```python
yearfilter2020df.to_csv("cleaned/2020clean.csv", index=False, encoding="UTF-8")
```
