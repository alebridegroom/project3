```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
census_2016 = "../CSV files/2016censustidy.csv"
election_2016 = "../CSV files/2016election.csv"
federal_2016 = "../CSV files/2016federal.csv"
state_2016 = "../CSV files/2016state.csv"

# Read the mouse data and the study results
census_2016 = pd.read_csv(census_2016 )
election_2016 = pd.read_csv(election_2016)
federal_2016 = pd.read_csv(federal_2016)
state_2016 = pd.read_csv(state_2016)


```
