```python
import pandas as pd
import sqlite3
```

```python
census_df = pd.read_csv("cleaned_2016/census_2016.csv")
```

```python
census_df
```

```python
conn = sqlite3.connect("election_db.sqlite")
```

```python
census_df.to_sql("2016_census", conn, if_exists= "replace", index=False)
```

```python
pd.read_sql("select * from '2016_census'", conn)
```

```python
election = pd.read_csv("cleaned_2016/election_2016.csv")
election
```

```python
election.to_sql("2016_election", conn, if_exists= "replace", index=False)
```

```python
pd.read_sql("select * from '2016_election'", conn)
```
