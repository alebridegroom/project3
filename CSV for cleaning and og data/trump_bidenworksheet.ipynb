```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Name of the CSV file
trump_bidencsv = Path('uncleaned/trump_biden_polls.csv')
```

```python
# The correct encoding must be used to read the CSV in pandas
trump_bidendf = pd.read_csv(trump_bidencsv, encoding="UTF-8")
```

```python
trump_bidendf.head()
```

```python
cleantrump_bidendf = trump_bidendf[['cycle', 'state', 'stage', 'answer', 'candidate_name', 'candidate_party', 'pct']]
cleantrump_bidendf.head()
```

```python
cleantrump_bidendf.to_csv("cleaned/trump_bidenclean.csv", index=False, encoding="UTF-8")
```
