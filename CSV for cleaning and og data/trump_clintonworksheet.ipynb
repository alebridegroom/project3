```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Name of the CSV file
trump_clintoncsv = Path('uncleaned/trump_clinton_polls.csv')
```

```python
# The correct encoding must be used to read the CSV in pandas
trump_clintondf = pd.read_csv(trump_clintoncsv, encoding="UTF-8")
```

```python
trump_clintondf.head()
```

```python
cleantrump_clintondf = trump_clintondf[['start_date', 'end_date', 'state', 'sample_size', 'population', 'Clinton', 'Trump']]
cleantrump_clintondf.head()
```

```python
cleantrump_clintondf.to_csv("cleaned/trump_clintonclean.csv", index=False, encoding="UTF-8")
```
